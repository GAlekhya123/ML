```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Online Retail.xlsx")
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data.info()
```

```python
data = data[pd.notnull(data["CustomerID"])]
```

```python
data.info()
```

```python
filtered_data = data[["Country","CustomerID"]].drop_duplicates()
```

```python
filtered_data.Country.value_counts()[:10].plot(kind = "bar")
```

```python
uk_data = data[data.Country == "United Kingdom"]
```

```python
uk_data.info()
```

```python
uk_data.describe()
```

```python
uk_data = uk_data[(uk_data["Quantity"]>0)]
```

```python
uk_data.info()
```

```python
uk_data = uk_data[["CustomerID","InvoiceNo","InvoiceDate","Quantity","UnitPrice"]]
```

```python
uk_data
```

```python
uk_data["Totalprice"] = uk_data["Quantity"]*uk_data["UnitPrice"]
```

```python
uk_data["InvoiceDate"].min(),uk_data[""]
```

```python
present  = dt.datetime(2011,12,10)
uk_data["InvoiceDate"] = pd.to_datetime(uk_data["InvoiceDate"])
```

```python
uk_data.head()
```

```python
rfm = uk_data.groupby("CustomerID").agg({"InvoiceDate":lambda date:(present-date.max()).days,
                                         "InvoiceNo":lambda num:len(num),
                                         "Totalprice":lambda price:price.sum()})
```

```python
rfm.columns
```

```python
rfm.columns = ["recency","frequency","monetry"]
```

```python
rfm["recency"]=rfm["recency"].astype(int)
```

```python
rfm.head()
```

```python
rfm["r_quartile"] = pd.qcut(rfm["recency"],4,["1","2","3","4"])
rfm["f_quartile"] = pd.qcut(rfm["frequency"],4,["4",'3',"2","1"])
rfm["m_quartile"] = pd.qcut(rfm["monetry"],4,["4","3","2","1"])
```

```python
rfm.head()
```

```python
rfm["RFM_Score"] = rfm.r_quartile.astype(str)+rfm.f_quartile.astype(str)+rfm.m_quartile.astype(str)
```

```python
rfm.head()
```

```python
rfm[rfm["RFM_Score"]=="122"].sort_values("monetry",ascending = False).head(500)
```

```python
rfm[rfm["RFM_Score"]=="111"].sort_values("monetry",ascending = False).head(50)
```
